```python
# python visualization libraries
import pandas as pd
import numpy as np

import math
import json


import plotly.express as px
import plotly.graph_objects as go

# for hierarchical data
import networkx as nx
```

```python
# ref: https://stackoverflow.com/questions/32400867/pandas-read-csv-from-url

url = 'https://raw.githubusercontent.com/lpp5139/ist526/main/department_data.csv'

# pandas call to read csv file 
df = pd.read_csv(url)

# quickly show the dataframe
df.head()
```

```python
# remove NaN with blank, otherwise plotly will be upset
df.fillna('', inplace = True)
# df.dropna(axis=0, inplace = True)
display(df.head())
```

```python
# adding a root color often helps

fig = px.treemap(
  df,   
  parents = 'Parent',
  names = 'Department/Item Name ',
  values='Weight'
  #color='Indent Level'
)

fig.update_layout(
  title_text="Treemap Diagram", 
  font_size=12,
  margin = dict(t=50, l=25, r=25, b=25)
)

fig.update_traces(
  root_color="lightgrey"
)

fig.show()
```
